# <font color='red'>  SGD  using linear regression using sckilt learn </font>

```python

def sklearn_sgd(alpha,eta0=0.01, power_t=0.25, n_iter=None, x_train=x_train):
    clf=SGDRegressor(alpha=alpha, eta0=eta0 , power_t=power_t , n_iter=n_iter)
    clf.fit(x_train, y_train)
    y_pred=clf.predict(x_test)
    #scatter plot
    plt.scatter(y_test,y_pred)
    plt.title('scatter plot between actual y and predicted y')
    plt.xlabel('actual point of  y')
    plt.ylabel('predicted point of y')
    plt.grid(linewidth=0.5)
    plt.show()
    
    print('***********************************************************')
    
    #kdeplot
    sgd_error=mean_squared_error(y_test,y_pred)
    print('mean sq error=', int(sgd_error))
    print('number of iteration=', n_iter)
    return clf.coef_, clf.intercept_, sgd_error      
```
